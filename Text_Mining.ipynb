```python
def select_random_rows(DF, n_rows, seed = None):
    random_rows = DF.sample(n = n_rows, random_state = seed)
    return random_rows

## Scans through sentences in a corpus and finds the ones including the target word.
def find_sentences_with_word(doc, target_word):
  nlp = spacy.load('en_core_web_sm')
  doc = nlp(doc)
  sentences = [sent for sent in doc.sents]

  target_sentences = []
  for sentence in sentences:
    if any(token.text.lower() == target_word.lower() for token in sentence):
            target_sentences.append(sentence.text)

  result = pd.DataFrame({'Sentences': target_sentences})
  return result

def find_sentences_from_docs(docs, target_word):
  Sentences = pd.DataFrame(columns = ['Sentences'])
  for doc in tqdm(docs, "Scanning documents"):
    results = find_sentences_with_word(doc, target_word)
    Sentences = pd.concat([Sentences, results], axis = 0).reset_index(drop=True)

  return Sentences
```
